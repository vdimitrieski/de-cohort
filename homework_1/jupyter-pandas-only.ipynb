```python
import pandas as pd
```

### Download neccessary data

```python
os.system('wget https://github.com/DataTalksClub/nyc-tlc-data/releases/download/green/green_tripdata_2019-01.csv.gz -O tripdata.csv.gz')
```

```python
os.system('wget https://s3.amazonaws.com/nyc-tlc/misc/taxi+_zone_lookup.csv -O zones.csv')
```

```python
df_trip = pd.read_csv('tripdata.csv.gz')
```

```python
df_trip.head()
```

```python
df_trip.lpep_pickup_datetime = pd.to_datetime(df_trip.lpep_pickup_datetime)
df_trip.lpep_dropoff_datetime = pd.to_datetime(df_trip.lpep_dropoff_datetime)
```

```python
df_zones = pd.read_csv('zones.csv')
```

```python
df_zones.head()
```

```python
df = df_trip.merge(df_zones, left_on='PULocationID', right_on='LocationID')
df = df.merge(df_zones, left_on='DOLocationID', right_on='LocationID', suffixes=('_PU', '_DO'))
```
